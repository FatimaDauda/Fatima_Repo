```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
df= pd.read_csv("C:\\Users\\Asus\\Desktop\\Python Class\\files\\SampleSuperstore.csv")
df.head(3)
```

```python
## adding a kde plot to the histplot and HUE
sb.set_theme(style="darkgrid")
sb.histplot(data=df,x=df["Postal Code"],bins=np.arange(0,100000,10000),
               kde= True,
               hue=df.Segment,
           palette="twilight")
```

```python
sb.set_theme(style="darkgrid")
sb.histplot(data=df,x=df["Postal Code"],
            hue=df.Segment,
           palette="gist_heat",
           shrink=0.5,
           element="poly")
```

#### Heat map:
graphical representation of a data,depicted by the density of a color 

```python
## Using correlation for heat map
df= pd.read_csv("C:\\Users\\Asus\\Desktop\\Python Class\\world_development_data_interpolated.csv")
df.head(3)
```

```python
aa=df.corr()
```

```python
sb.heatmap(data=aa)
plt.show()
```

```python
sb.heatmap(data=aa,
          vmin=-1,
          vmax=1,
          cmap="RdBu")
plt.show()
```

```python
## Using pivot table for heat map
df= pd.read_csv("C:\\Users\\Asus\\Desktop\\Python Class\\Auto Sales data.csv")
df.head(3)
```

```python
##Creating a pivot table
a= pd.pivot_table(data=df,
              columns="PRODUCTLINE",
              index='COUNTRY',
              values='SALES',
              aggfunc=np.sum)
```

```python
sb.heatmap(data=a,cmap="flare_r",
          square=True)
plt.show()
```

```python
sb.heatmap(data=a,cmap="flare",
          #square=True,
          annot=True,
          fmt=".0f",
          linewidths=1,
          linecolor="k")
plt.show()
```

```python
sb.heatmap(data=a,cmap="flare",
          #square=True,
          annot=True,
          fmt=".0f",
          linewidths=0.6,
          linecolor="k",
          annot_kws=dict(size=10,weight= "bold"))
plt.show()
```
